```python
import cv2
import numpy as np
from numpy.core.shape_base import hstack


"""改變飽和度"""
def change_saturation(change_percentage: float):
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    img_hsv_change_s = img_hsv.astype("float32")
    img_hsv_change_s[:, :, 1] = img_hsv_change_s[:, :, 1] / 255 + change_percentage
    # 設定邊界
    for row in img_hsv_change_s[..., 1]:
        for pixel in row:
            if pixel > 1:
                pixel = 1
            elif pixel < 0 :
                pixel = 0

    img_hsv_change_s[:, :, 1] = img_hsv_change_s[:, :, 1] * 255
    img_hsv_change_s = img_hsv_change_s.astype("uint8")
    img_new = cv2.cvtColor(img_hsv_change_s, cv2.COLOR_HSV2BGR)
    concatenate = np.hstack([img, img_new])
    cv2.imshow("Change saturation", concatenate)
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()


"""對彩圖的每個channel做直方圖均衡"""
def histogram_equalization_BGR():
    for channel in range(img.shape[2]):
        imgHE = cv2.imread("lena.png")
        imgHE[..., channel] = cv2.equalizeHist(imgHE[..., channel])
    
    concatenate = np.hstack([img, imgHE])
    cv2.imshow("test", concatenate)

    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()


"""對HSV的其中一個channel做直方圖均衡"""
def histogram_equalization_HSV():
    imgHE = cv2.imread("lena.png")
    imgHE = cv2.cvtColor(imgHE, cv2.COLOR_BGR2HSV)
    imgHE[..., -1] = cv2.equalizeHist(imgHE[..., -1])
    
    concatenate = np.hstack([img, imgHE])
    cv2.imshow("test", concatenate)

    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()

"""調整對比/明亮"""
def adjust_contrast_lightness():
    add_contrast = cv2.convertScaleAbs(img, alpha=2.0, beta=0)
    add_lightness = cv2.convertScaleAbs(img, alpha=1.0, beta=50)

    concatenate = np.hstack([img, add_contrast, add_lightness])
    cv2.imshow("Adjust contrast/lightness", concatenate)
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()


if __name__ == "__main__":
    img = cv2.imread("lena.png", cv2.IMREAD_COLOR)
    # change_saturation(-0.2)
    # histogram_equalization_BGR()
    # histogram_equalization_HSV()
    adjust_contrast_lightness()
```
